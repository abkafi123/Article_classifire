```python
import matplotlib.pyplot as plt
import numpy as np 
```

### Dimension

```python
abstract_dim = [76.84,80.46,79.68,76.35,75.35]
authorkeyword_dim = [67.03,67.25,66.03,66.0,65.64]
articletitle_dim = [65.8,66.35,64.63,66.02,64.55]
labels = ['NB', 'LR', 'SVM', 'CNN', 'ANN']
```

```python
x = np.arange(len(labels))  # the label locations
y = np.arange(10,100,10)   
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(10)

rects1 = ax.bar(x - width/2, abstract_dim, width, label='Abstract')
rects2 = ax.bar(x + width/2, authorkeyword_dim, width, label='AuthorKeyword')
rects3 = ax.bar(x + width*1.5, articletitle_dim, width, label='ArticleTitle')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores',size=10,fontweight='bold')
ax.set_title('Accuracy (Dimension)', size=14,fontweight='bold')
ax.set_xlabel('Algorithms',size=10,fontweight='bold')
ax.set_xticks(x, labels)
ax.set_yticks(y)
ax.legend()

ax.bar_label(rects1, padding=2)
ax.bar_label(rects2, padding=2)
ax.bar_label(rects3, padding=2)

fig.tight_layout()

plt.show()
```
